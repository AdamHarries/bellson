```python
import glob
import re 
# Hardcode a path to a directory
rundir = "/mnt/bigboi/training_runs/2020-05-22_10.00"
name_regex = "model-epoch-(\d+)-loss-(\d\.\d+).hdf5"
```

```python
[m for m in glob.glob(rundir + "/*.hdf5")]
```

```python
latest_epoch = int(0)
latest_model_file = ""
```

```python
for model_file in glob.glob(rundir + "/*.hdf5"): 
    try: 
        m = re.search(name_regex, model_file)
        epoch = int(m.group(1))
        loss = float(m.group(2))
        print(f"Epoch: {epoch}") 
        print(f"Loss: {loss}")
        if epoch > latest_epoch: 
            latest_epoch = epoch 
            latest_model_file = model_file
            print(f"New latest model file: {model_file}")
    except AttributeError: 
        print("Some error")
```
